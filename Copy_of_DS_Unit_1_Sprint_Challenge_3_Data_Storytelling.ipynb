```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

url = 'https://raw.githubusercontent.com/fivethirtyeight/data/master/daily-show-guests/daily_show_guests.csv'
df = pd.read_csv(url).rename(columns={'YEAR': 'Year', 'Raw_Guest_List': 'Guest'})

def get_occupation(group):
    if group in ['Acting', 'Comedy', 'Musician']:
        return 'Acting, Comedy & Music'
    elif group in ['Media', 'media']:
        return 'Media'
    elif group in ['Government', 'Politician', 'Political Aide']:
        return 'Government and Politics'
    else:
        return 'Other'
      
df['Occupation'] = df['Group'].apply(get_occupation)
```

# Part 1 — What's the breakdown of guests’ occupations per year?

For example, in 1999, what percentage of guests were actors, comedians, or musicians? What percentage were in the media? What percentage were in politics? What percentage were from another occupation?

Then, what about in 2000? In 2001? And so on, up through 2015.

So, **for each year of _The Daily Show_, calculate the percentage of guests from each occupation:**
- Acting, Comedy & Music
- Government and Politics
- Media
- Other

#### Hints:

You'll know you've calculated correctly when the percentage of "Acting, Comedy & Music" guests is 90.36% in 1999, and 45% in 2015.

```python
df['counter'] = 1
```

```python
#first we try crosstab on 1999 before generalising
df_1999 = df[df['Year']==1999]
ct = pd.crosstab(df_1999['Occupation'],df_1999['counter']).head()
ct = ct / list(ct.sum())[0]
ct.columns = [1999]
#ct[1999]] = pd.Series(["{0:.2f}%".format(val * 100) for val in ct[1999]], index = ct.index)
ct.head(10)
```

```python
#that seems to be workign so we generalise with a pivot

pivot = df.pivot_table(index='Occupation',columns='Year',values='counter',aggfunc=sum)
for column in pivot.columns:
  pivot[column] = pivot[column]/pivot[column].sum()

pivot
```

```python
import seaborn as sns
sns.set()
ax = pivot.T.plot(kind='bar', stacked=True, colormap = 'Spectral')
ax.figure.set_size_inches(15, 8)
ax.set(facecolor='#f9f9f9')
ax.set_title('Stacked Bar Chart - Guest of the Daily Show - % of occupation by year')
ax.set_ylim(0,1.4);
```

```python
# we can see the Daily show has gone from interviewing pretty much only actors to being more balanced, 
# though % has risen in recent years at the expense of government
```

```python
ct_guest.columns = ['appearances']
#let's cut out the people who only appeared 9 times or less
ct_guest = ct_guest[ct_guest['appearances']>9]
ct_guest = ct_guest.sort_values(by='appearances',ascending=False)
ct_guest.head(15)
```

```python
fig2, ax2 = plt.subplots()
y = ct_guest.index
x = ct_guest['appearances']

width = 0.75
ax2.barh(y, x, width, align='center', color='#87CEEB', ecolor='black')
ax2.figure.set_size_inches(10, 10)
ax2.invert_yaxis()
ax2.set_title('Who were the top 10 guests on The Daily Show?',fontsize=20,fontweight='bold')

x_ticks = [5,10,15,20]
plt.xticks(ticks=x_ticks,fontsize=15)
plt.yticks(fontsize=15)
ax2.xaxis.set_label_coords(0.4,0.97)

xvals = list(x)
for i, v in enumerate(xvals):
    ax2.text(v + 0.25, i + .25, str(v), color='k', fontweight='bold')
ax2.set(facecolor='#f9f9f9')
ax2.set_ylim(bottom='Bill Maher',top='Fareed Zakaria',auto=True)

plt.margins(0.001);
```
