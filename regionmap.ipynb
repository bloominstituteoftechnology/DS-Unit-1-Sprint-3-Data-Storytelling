```
#mapping us regions to states - i copied this from wikipedia and had to manually adjust it!  i'm sure there was a better way!

New_England = ['New_England','Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont']
Mid_Atlantic =  ['Mid_Atlantic','New Jersey', 'New York','Pennsylvania']
East_North_Central = ['East_North_Central','Illinois', 'Indiana', 'Michigan', 'Ohio', 'Wisconsin']
West_North_Central = ['West_North_Central','Iowa', 'Kansas', 'Minnesota', 'Missouri', 'Nebraska', 'North Dakota','South Dakota']
South_Atlantic =['South_Atlantic','Delaware', 'Florida', 'Georgia', 'Maryland', 'North Carolina', 'South Carolina', 'Virginia', 'District of Columbia', 'West Virginia']
East_South_Central = ['East_South_Central','Alabama', 'Kentucky', 'Mississippi','Tennessee']
West_South_Central = ['West_South_Central','Arkansas', 'Louisiana', 'Oklahoma', 'Texas']
Mountain = ['Mountain','Arizona', 'Colorado', 'Idaho', 'Montana', 'Nevada', 'New Mexico', 'Utah','Wyoming']
Pacific = ['Pacific','Alaska', 'California', 'Hawaii', 'Oregon', 'Washington']
```

```
region_list = [New_England,Mid_Atlantic,East_North_Central,West_North_Central,South_Atlantic,East_South_Central,West_South_Central,Mountain,Pacific]
```

```
region_dict={}
for region in region_list:
  for i in range(1,len(region)):
    region_dict[region[i]] = region[0]
    
region_dict
```

```
region_dict.get('North Dakota')
```

```

import pandas as pd
import numpy as np
url = 'https://raw.githubusercontent.com/LambdaSchool/DS-Sprint-02-Storytelling-With-Data/master/module2-choose-appropriate-visualizations/direct_written_premium_by_state_by_group_private_passenger_auto.csv'
df = pd.read_csv(url)
```

```
df_state = df[['state','id']].copy()
df_state.head()
```

```
df_state['region'] = [region_dict.get(w) for w in df['state']]
```

```
df_state = df_state.dropna()
```

```
df_state.isna().sum()
```

```
df_state['region'] = pd.Categorical(df_state['region'])
#df_state['region_code'] = df_state.region.categories.codes
  
```

```
df_state['region_code'] = df_state['region'].cat.codes
```

```
df_state.head(51)
```

```
region_domain=['East_North_Central','East_South_Central','Mid_Atlantic','Mountain','New_England','Pacific','South_Atlantic','West_North_Central','West_South_Central']
```

```
import altair as alt
import matplotlib.pyplot as plt
from vega_datasets import data


states = alt.topo_feature(data.us_10m.url, 'states')
variable = 'region'

alt.Chart(states).mark_geoshape().encode(
    alt.Color(variable, type='nominal')
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(df_state, 'id', [variable])
).properties(
    width=500,
    height=300
).project(
    type='albersUsa'
)

```

```
sides_dict={0:'ROLLS/BISCUITS',1:'MAC & CHEESE',2:'SQUASH',3:'SALAD',4:'SQUASH',5:'SALAD',6:'MAC & CHEESE',7:'GREEN BEANS/CASSEROLE',8:'CORNBREAD'}
```

```
df_state['favourite side'] = [sides_dict.get(w) for w in df_state['region_code']]
```

```
df_state
```

```
sides_domain = ['CORNBREAD','GREEN BEANS/CASSEROLE','MAC & CHEESE','ROLLS/BISCUITS','SALAD','SQUASH']
sides_colors = ['#F4A460','#FFD700','#6A5ACD','#32CD32','#20B2AA','#EE82EE']
```

```
states = alt.topo_feature(data.us_10m.url, 'states')
variable = 'favourite side'

alt.Chart(states).mark_geoshape().encode(
    alt.Color(variable, type='nominal',scale=alt.Scale(domain=sides_domain, range=sides_colors))
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(df_state, 'id', [variable])
).properties(
    width=500,
    height=300
).project(
    type='albersUsa'
)
```
