```python
sns.relplot('x', 'y', col='dataset', data=df, col_wrap=2);
```

```python
sns.lmplot('x', 'y', col='dataset', hue='dataset', col_wrap=2, ci=None, data=df);
```

## 2. Tips dataset

### Load dataset

```python
tips = sns.load_dataset('tips')
```

### See the data's shape

```python
tips.shape
```

### See the first 5 rows

```python
tips.head()
```

### Describe the data

```python
tips.describe()
```

```python
import numpy as np
tips.describe(exclude=np.number)
```

### Make univariate [distribution plots](https://seaborn.pydata.org/generated/seaborn.distplot.html)

```python
sns.distplot(tips.tip);
```

```python
# Making column that shows tip as percent of total bill
tips['percent'] = tips['tip'] / tips['total_bill']
```

```python
# Seeing new column
tips.head()
```

```python
sns.distplot(tips.percent);
```

### Make bivariate [relational plots](https://seaborn.pydata.org/generated/seaborn.relplot.html)

```python
sns.relplot(x='total_bill', y='percent', data=tips, alpha=0.5);
```

```python
x='total_bill'
y='tip'

sns.relplot(x, y, data=tips, alpha=0.5)
```

### Make univariate [categorical plots](https://seaborn.pydata.org/generated/seaborn.catplot.html)

```python
# Plotting with pandas
tips.sex.value_counts().plot.bar();
```

```python
# Plotting with seaborn
sns.catplot('sex', data=tips, kind='count');
```

### Make bivariate [categorical plots](https://seaborn.pydata.org/generated/seaborn.catplot.html)

```python
sns.catplot(x='sex', y='tip', data=tips, kind='strip', alpha=0.5);
```

```python
sns.catplot(x='sex', y='tip', data=tips, kind='box');
```

```python
sns.catplot(x='sex', y='tip', data=tips, kind='bar');
```

```python
# This is a confounding variable because it influences the size of the tip
sns.catplot(x='sex', y='total_bill', data=tips, kind='bar');
```

```python
sns.catplot(x='sex', y='percent', data=tips, kind='bar');
# There's not enough evidence to say women tip more than men or vice versa
```

```python
sns.heatmap(flights_piv);
```
