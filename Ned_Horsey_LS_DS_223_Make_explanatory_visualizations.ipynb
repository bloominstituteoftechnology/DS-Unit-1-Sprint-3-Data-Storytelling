## Annotate with text

```python
plt.style.use('fivethirtyeight')
# fig = plt.fig
fake = pd.Series([38, 3, 2, 1, 2, 4, 6, 5, 5, 33], 
                 index=range(1,11))

ax = fake.plot.bar(color='#f58120', width=0.9);

ax.set(yticks=range(0,50,10), facecolor='white'
      )
# plt.title("'An Inconvenient Sequel: Truth To Power' is divisive")
plt.xlabel('Rating', fontsize=10)
plt.ylabel('Percentage of Total Votes',fontsize=10 )
ax.text(x=-1.3, y=44, s="'An Inconvenient Sequel: Truth To Power' is divisive", 
        fontsize=14, fontweight='bold' );
ax.text(x=-1.2, y=42, s="IMDB ratings for the film as of Aug. 29", 
        fontsize=10);

ax.tick_params(labelrotation=0)
# ax.set_facecolor('xkcd:salmon')
```

```python
lastday = df['2017-08-29']
```

```python
lastday[lastday['category'] == 'IMDb users'].respondents.plot()
```

```python
final = lastday[lastday['category'] == 'IMDb users' ].tail(1)
```

```python
final
```

```python
pct_columns = ['1_pct', '2_pct', '3_pct', '4_pct', '5_pct',
                '6_pct', '7_pct', '8_pct', '9_pct', '10_pct']

final[pct_columns]
```

```python
data = final[pct_columns].T

data.index = range(1,11)

data
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure();
fig.patch.set_facecolor('#f58120')
fake = pd.Series([38, 3, 2, 1, 2, 4, 6, 5, 5, 33], 
                 index=range(1,11))


ax = data.plot.bar(color='#f58120', width=0.9);

ax.set(yticks=range(0,50,10), facecolor='white'
      )
# plt.title("'An Inconvenient Sequel: Truth To Power' is divisive")
plt.xlabel('Rating', fontsize=10)
plt.ylabel('Percentage of Total Votes',fontsize=10 )
ax.text(x=-1.3, y=44, s="'An Inconvenient Sequel: Truth To Power' is divisive", 
        fontsize=14, fontweight='bold' );
ax.text(x=-1.2, y=42, s="IMDB ratings for the film as of Aug. 29", 
        fontsize=10);
plt.gca().get_legend().remove()
ax.tick_params(labelrotation=0)
# ax.set_facecolor('xkcd:salmon')
```
