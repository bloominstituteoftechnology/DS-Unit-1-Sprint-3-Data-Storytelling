## Annotate with text

```python
plt.style.use('fivethirtyeight')
fig = plt.figure()
fake = pd.Series([38, 3, 2, 1, 2, 4, 6, 5, 5, 33], 
                 index=range(1,11))

ax = fake.plot.bar(color='#ef7030', width=0.9)

ax.set(yticks=range(0, 50, 10),
       facecolor='#f9f9f9')
#ax.set_axis_bgcolor('#f9f9f9')
#ax.set_facecolor('white')
fig.patch.set_facecolor('#f9f9f9')

#plt.title("'An Inconvenient Sequel: Truth to Power' is divisive")
#plt.ylabel('Percent of total votes')
#plt.xlabel('Rating')
plt.ylabel('Percent of Total Votes', fontsize=10, fontweight='bold')
plt.xlabel('Rating', fontsize=10, fontweight='bold')

ax.text(x=-2, y=45, s="'An Inconvenient Sequel: Truth to Power' is divisive",
       fontsize=14, fontweight='bold')
ax.text(x=-2, y=42, s="IMDB ratings for the film as of Aug. 29",
       fontsize=12)
ax.tick_params(labelrotation=0)
```

```python
df.set_index('timestamp', inplace=True)
```

```python
df['2017-08-29']
```

```python
df['category'].value_counts()
```

```python
lastday = df['2017-08-29']
```

```python
lastday[lastday['category'] == 'IMDb users']
```

```python
lastday[lastday['category'] == 'IMDb users'].respondents.plot()
```

```python
final = lastday[lastday['category'] == 'IMDb users'].tail(1)
```

```python
final
```

```python
pct_columns = ['1_pct', '2_pct', '3_pct', '4_pct', '5_pct',
               '6_pct', '7_pct', '8_pct', '9_pct', '10_pct']
final[pct_columns]
```

```python
data = final[pct_columns].T
data.index = range(1, 11)
data
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure();
fig.patch.set_facecolor('#f9f9f9');

ax = data.plot.bar(color='#ef7030', width=0.9)

ax.set(yticks=range(-3, 50, 10),
       facecolor='#f9f9f9');
#ax.set_axis_bgcolor('#f9f9f9')
#ax.set_facecolor('white')

#plt.title("'An Inconvenient Sequel: Truth to Power' is divisive")
#plt.ylabel('Percent of total votes')
#plt.xlabel('Rating')
plt.ylabel('Percent of Total Votes', fontsize=10, fontweight='bold');
plt.xlabel('Rating', fontsize=10, fontweight='bold');
ax.get_legend().remove()

ax.text(x=-2, y=55, s="'An Inconvenient Sequel: Truth to Power' is divisive",
       fontsize=14, fontweight='bold');
ax.text(x=-2, y=52, s="IMDB ratings for the film as of Aug. 29",
       fontsize=12);
ax.tick_params(labelrotation=0);
```

```python
import altair as alt
from vega_datasets import data

source = data.cars()

alt.Chart(source).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='Origin:N'
).properties(
    width=150,
    height=150
).repeat(
    row=['Horsepower', 'Acceleration', 'Miles_per_Gallon'],
    column=['Miles_per_Gallon', 'Acceleration', 'Horsepower']
).interactive()
```

```python
url='https://raw.githubusercontent.com/fivethirtyeight/data/master/candy-power-ranking/candy-data.csv'
df = pd.read_csv(url)
```

```python
df.shape
```

```python
pd.set_option('display.max_columns', 500)
df.head(8)
```

```python
df['rank'] = df['winpercent'].rank(ascending=False, method='max')
```

```python
du=df.set_index('rank').sort_index(ascending)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
#https://fivethirtyeight.com/features/the-ultimate-halloween-candy-power-ranking/
data=[go.Bar(
y=du[['competitorname']].sort_index(ascending=False),
    width=.5,
x=du['winpercent'].sort_index(ascending=False),
marker=dict(
  color='rgb(244, 92, 66)'),
orientation='h',
name='Candies'
)]
layout = go.Layout(
  autosize=False,
  width=500,
  height=2000,
  title='How often did a fun-sized candy of a given type win its matchups against the rest of the field?',
  xaxis=dict(
    tickfont=dict(
      size=20,
      color='rgb(107,107,107)'
    )
  ),
  yaxis=go.layout.YAxis(
    title='Y-title',
    automargin=True))
fig = go.Figure(data=data,layout=layout)
py.iplot(fig, filename='basic')
```
